```python
import  pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1',headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
company = soup.find_all('div', class_ = 'company-content')
```

```python
name = []
rating = []
reviews = []
ctype = []
hq = []
old = []
employee = []

for i in company:

    name.append(i.find('h2').text.strip())

    rating.append(i.find('p',class_='rating').text.strip())

    reviews.append(i.find('a', class_ = "review-count").text.strip())

    ctype.append(i.find_all('p', class_ = 'infoEntity')[0].text.strip())

    hq.append(i.find_all('p', class_ = 'infoEntity')[1].text.strip())

    old.append(i.find_all('p', class_ = 'infoEntity')[2].text.strip())

    try:
        employee.append(i.find_all('p', class_ = 'infoEntity')[3].text.strip())
    except:
        employee.append(np.nan)    
    
```

```python
d = {'name':name, 'rating':rating, 'reviews':reviews, 'comp_type':ctype, 'head_quarter':hq, 'how_old':old, 'no_employee':employee}

df = pd.DataFrame(d)
df
```

```python
final = pd.DataFrame()

for j in range(1,331):
    url = f"https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={j}"
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url,headers=headers).text

    soup = BeautifulSoup(webpage,'lxml')
    company = soup.find_all('div', class_ = 'company-content')

    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []
    old = []
    employee = []

    for i in company:
      try:
        name.append(i.find('h2').text.strip())
      except:
        name.append(np.nan)  
      
      try:
        rating.append(i.find('p',class_='rating').text.strip())
      except:
        rating.append(np.nan)  
      
      try:
        reviews.append(i.find('a', class_ = "review-count").text.strip())
      except:
        reviews.append(np.nan)
      
      try:
        ctype.append(i.find_all('p', class_ = 'infoEntity')[0].text.strip())
      except:
        ctype.append(np.nan)
      
      try:
        hq.append(i.find_all('p', class_ = 'infoEntity')[1].text.strip())
      except:
        hq.append(np.nan)
      
      try:
        old.append(i.find_all('p', class_ = 'infoEntity')[2].text.strip())
      except:
        old.append(np.nan)

      try:
        employee.append(i.find_all('p', class_ = 'infoEntity')[3].text.strip())
      except:
        employee.append(np.nan)  
          
    d = {'name':name, 'rating':rating, 'reviews':reviews, 'comp_type':ctype, 'head_quarter':hq, 'how_old':old, 'no_employee':employee}

    df = pd.DataFrame(d)  
    final = final.append(df, ignore_index=True)
```
