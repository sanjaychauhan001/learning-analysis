```python
def male_female_child(passenger):
    age, sex = passenger
    if age < 16:
        return "child"
    else:
        return sex
```

```python
titanic_df["person"] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
deck = titanic_df["Cabin"].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    

cabin_df = DataFrame(levels)

cabin_df.columns = ["Cabin"]

sns.countplot(x=cabin_df["Cabin"],palette='winter_d')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']

sns.countplot(x=cabin_df["Cabin"],palette='summer',order=['A','B','C','D','E','F','G'])
```
