```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_csv(r"C:\Users\\sanja\\Downloads\\100-days-of-machine-learning-main\\day48-simple-linear-regression\\placement.csv")
```

```python
sns.scatterplot(x=df['cgpa'], y=df['package'])
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
my_model = LinearRegression()
```

```python
my_model.fit(X_train, y_train)
```

```python
predictions = my_model.predict(X_test)
```

```python
r2_score(y_test, predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))# model is making error of 34 thousands
```

```python
sns.regplot(data=df, x='cgpa', y='package') # this is our best fit line
```

```python
residuals = y_test - predictions
```

```python
sns.kdeplot(residuals) # assumption residuals are normally distributed
```
