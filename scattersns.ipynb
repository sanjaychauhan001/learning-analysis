```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("penguins")
data.head()
```

```python
m ={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=data,hue="sex",
                size="sex",sizes=(40,20),palette="gist_ncar",alpha=1,
                markers=m)
plt.show()
```
