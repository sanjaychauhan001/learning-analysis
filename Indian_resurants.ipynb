```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\sanja\\Downloads\\restaurants.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
Top_10_rsturants = df.Name.value_counts().sort_values(ascending = False).head(10)
Top_10_rsturants = Top_10_rsturants.reset_index()
Top_10_rsturants.rename(columns={'index':"resturant","Name":'frequency'},inplace=True)
Top_10_rsturants
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='resturant',y='frequency',data=Top_10_rsturants)
```

```python
top_resturant_rating = df.groupby('Name')['Rating'].sum().sort_values(ascending=False).head(10)
top_resturant_rating = top_resturant_rating.reset_index()
top_resturant_rating
```

```python
plt.figure(figsize=(12,5))

ax = sns.barplot(x="Name",y="Rating",data=top_resturant_rating)

plt.title("Top Resturant By Rating")
plt.show()
```
