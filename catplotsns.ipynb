```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.catplot(x="size",y="tip",data=data,kind="point")
```

```python
sns.catplot(x="day",y="total_bill",data=data,hue="sex",kind="bar")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=data,hue="sex",kind="box")
plt.show()
```
