```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
files = [file for file in os.listdir("C:\\Users\\sanja\\Downloads\\C5 Input for participants\\domestic_visitors") ]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("C:\\Users\\sanja\\Downloads\\C5 Input for participants\\domestic_visitors\\"+file)
    all_months_data = pd.concat([all_months_data,df])

all_months_data.to_csv("C:\\Users\\sanja\\Downloads\\C5 Input for participants\\domestic_visitors.csv",index=False)
```
