```python
sns.scatterplot(data=df, x="speed_mph", y="Height_ft",hue="Year_Introduced")
```

```python
sns.pairplot(df, vars=["Year_Introduced","speed_mph","Height_ft","Inversions","Gforce"],\
             hue="Type_Main")
plt.show()
```

```python
df_corr = df[["Year_Introduced","speed_mph","Height_ft","Inversions","Gforce"]].dropna().corr()
```

```python
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
```

```python
ax = df.query('Location != "Other"').groupby("Location")["speed_mph"]\
.agg(['mean','count']).query('count >=10').sort_values('mean')['mean']\
.plot(kind="barh", figsize=(12,5),title="Average coast speed by Location")
ax.set_xlabel("Average coaster speed")
plt.show()
```
