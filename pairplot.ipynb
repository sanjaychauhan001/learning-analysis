```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.pairplot(data,hue="sex")
plt.show()
```

```python
sns.pairplot(data,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(data,x_vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(data,y_vars=["total_bill","tip"],hue="sex")
plt.show()
```
