```python
deliveries = pd.read_csv("C:\\Users\\sanja\\Downloads\\deliveries.csv")
matches = pd.read_csv("C:\\Users\\sanja\\Downloads\\matches.csv")

course = pd.read_csv("C:\\Users\\sanja\\Downloads\\courses.csv")
nov = pd.read_csv("C:\\Users\\sanja\\Downloads\\reg-month1.csv")
dec = pd.read_csv("C:\\Users\\sanja\\Downloads\\reg-month2.csv")
students = pd.read_csv("C:\\Users\\sanja\\Downloads\\students (2).csv")
```

```python
regs = pd.concat([nov,dec],ignore_index=True)
regs.head()
```

```python
multi = pd.concat([nov,dec],keys=['Nov','Dec'])
multi.loc['Nov']
```

```python
multi.loc[('Nov',2)]
```

```python
students.merge(regs,how='inner',on='student_id')
```

```python
# find total revenue
regs.merge(course,how='inner',on='course_id')['price'].sum()
```

```python
# revenue by month
temp_df = pd.concat([nov,dec],keys=['NOV','DEC']).reset_index()
temp_df.merge(course,how='inner',on='course_id').groupby('level_0')['price'].sum()
```
