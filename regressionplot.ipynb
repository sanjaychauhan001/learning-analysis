```python
import pandas as pd
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
tips = pd.read_clipboard()
```

```python
tips.head()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip",
           scatter_kws={'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
tips['tip_pect'] = 100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot(data=tips,x='size',y='tip_pect')
```

```python
sns.lmplot(data=tips,x='size',y='tip_pect',x_estimator=np.mean)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip_pect',hue="sex",markers=['*','o'])
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip_pect',hue="day")
```
