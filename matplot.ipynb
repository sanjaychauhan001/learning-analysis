```python
days = [1,2,3,4,5,6,7]
max_t = [50,62,56,49,59,53,55]
min_t = [41,46,43,40,48,43,45]
avg_t = [45.5, 54. , 49.5, 44.5, 53.5, 48. , 50.]
```

```python
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather")

plt.plot(days,max_t, label = "max")
plt.plot(days, min_t,label = "min")
plt.plot(days, avg_t, label = "avg")

plt.legend()
plt.grid()
```

```python
import numpy as np
```

```python
company = ["google",'microsfot','apple','meta']

revenue = [136,100,150,95]

profit = [50,40,30,55]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue, label = "Revenue", width= 0.4)
plt.bar(xpos+0.2, profit, label = "Profit", width= 0.4)
plt.ylabel("revenue(bln)")
plt.title("us tech stocks")
plt.legend()
```

```python
blood_sugar_men = [113,111,96,100,105,115,90,103,96,99,108,101,80,86,126]
blood_sugar_women = [111,120,113,90,105,80,115,111,126,89,130]
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins = 3, rwidth=0.95,
         color= ["red","green"], label=["men","women"])

plt.xlabel("sugar range")
plt.ylabel("frequency")
plt.title("blood sugar analysis")
plt.legend()
```
