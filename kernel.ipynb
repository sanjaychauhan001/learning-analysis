```python
import pandas as pd
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)
plt.ylim(0,1)
```

```python
plt.hist(dataset, alpha = 0.3)
sns.rugplot(dataset)
```

```python
sns.kdeplot(dataset)
```

```python
sns.rugplot(dataset, color = "black")

for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset, bw_method  = bw, lw =0.8, label=bw)

plt.legend()    
```
