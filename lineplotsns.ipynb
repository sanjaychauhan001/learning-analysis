```python
import pandas as pd
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [3,4,5,4,5,2,4]
```

```python
x_1 = pd.DataFrame({"var":var, "var_1":var_1})
x_1
```

```python
sns.lineplot(x = var, y = var_1, data=x_1)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins").head(30)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex")
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",
             style="sex",palette="ocean",markers=["o",'D'])
plt.title("seaborn")
plt.show()
```
