```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
data
```

```python
data1 = data.groupby("day").sum()
data1
```

```python
sns.violinplot(x="day",y="total_bill",data=data)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data,linewidth=3,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=data,order=["Dinner",'Lunch'],inner="stick")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data,hue="sex")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data,hue="sex",split=True)
plt.show()
```

```python
sns.violinplot(x=data['total_bill'])
```
